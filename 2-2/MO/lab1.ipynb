```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
def f(x):
    return (1 / 8) * np.cos(x - 8)
# print(f(2))
```

```python
def tabulate_function(n = 21, p = 8):
    x0 = p
    xn = p + 7
    h = (xn - x0) / n
    x = [x0 + i * h for i in range(n + 1)]
    y = [f(xi) for xi in x]

    for xi, yi in zip(x, y):
        print(f"x: {xi:.6f}, y: {yi:.6f}")

    return x, y
```

```python
def compute_spline(x, y):
    cs = CubicSpline(x, y, bc_type='natural')
    return cs
```

```python
def plot_spline(x, y, cs):
    x_dense = np.linspace(x[0], x[-1], 1000)
    y_dense = cs(x_dense)

    plt.plot(x, y, 'ro', label='Табличні точки')
    plt.plot(x_dense, y_dense,  label='Кубічний сплайн')
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Кубічний сплайн інтерполяції")
    plt.show()

```

```python
x, y = tabulate_function()
cs = compute_spline(x, y)
plot_spline(x, y, cs)
```
